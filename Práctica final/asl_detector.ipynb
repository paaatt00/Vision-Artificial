# Creación del modelo


En primer lugar, contruiremos el dataset utilizando la clase `DataBlock`, que hará que los datos pasados por dicho objeto sean usados como datos de entrenamiento y de test.
* En este primer argumento, se especifica cuáles son las variables independientes y cuáles las dependientes. Aplicado a la detección del ASL, la variable independiente son las imágenes y la dependiente son las categorías, es decir, las letras.
```python
blocks=(ImageBlock, CategoryBlock)
```
* Fastai necesita una lista de los elementos que formarán el dataset. La función `get_image_files` retorna una lista de imágenes de una ruta dada.
```python
get_items=get_image_files
```
* El *splitter* separará el *training set* y el *validation set* con un `valid_pct` aleatoriamente. Para que en cada ejecución del programa se tengan los mismos sets se utiliza una semilla, `seed`, aleatoria.
```python
splitter=RandomSplitter(valid_pct=0.2, seed=15)
```
* A la variable independiente habitualmente se le llama `x` y a la variable dependiente `y`. En este argumento se le indica a fastai a qué función llamar para etiquetar nuestro dataset. `parent_label` es una función de fastai que coge el nombre de la carpeta en la que está la imagen.
```python
get_y=parent_label
```
* Se necesita que las imágenes sean del mismo tamaño, por eso añadimos una tranformación que aplica un reescalado a todas las imágenes.
```python
item_tfms=Resize(128)
```

Todo esto nos devuelve un objeto `DataBlock`, que será el *template* para la creación de un `DataLoaders`.



````python
hands = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=15),
    get_y=parent_label,
    item_tfms=Resize(128))
````

De forma predeterminada, `Resize` recorta las imágenes para que se ajusten a una forma cuadrada del tamaño dado, lo cual puede resultar en la pérdida de algunos detalles importantes de la imagen. Por eso, fastai incluye funciones para rellenar las imágenes con ceros (negro), para estirarlas o aplastarlas.

En el caso de las imágenes que se han utilizado para este dataset no sería necesario aplicar esto ya que, en principio, todas son cuadradas. Aun así, por si acaso, se aplicará la función de rellenar las imágenes con ceros.

````python
# hands = hands.new(item_tfms=Resize(128, ResizeMethod.Squish))
# hands = hands.new(item_tfms=RandomResizedCrop(128, min_scale=0.7)) # se coge el 70% de los píxeles
# hands = hands.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'), batch_tfms = [RGB2GreyTransform()]) # rellena de negro
# hands = hands.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=4))
hands = hands.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'))
````

Un `DataLoaders` incluye el *validation set* y el *training set*. `DataLoaders` es una clase que genera conjuntos, *batches*, de varios elementos a la vez a la GPU. 

````python
dls = hands.dataloaders('/home/patri/Documentos/ASLdetector/images')
````

````python
dls
````

````python
dls.valid.show_batch(max_n=8, nrows=2)
````

Ahora se creará el `Learner`. Se utilizará una red neuronal convolucional (CNN) siendo los parámetros:
* El objeto `DataLoaders` creado anteriormente, *dls*.
* La arquitectura que se va a utilizar, una `resnet50` en este caso.
* Las métricas que se desean mostrar, habiendo sido elegidas `accuracy` y `error_rate`.

````python
learn = cnn_learner(dls, resnet50, metrics=[accuracy, error_rate])
````

La función `freeze()` 'congela' los pesos y el bias de la red utilizada para que solo se modifiquen los pesos de las capas finales de la red. Así se conserva el conocimiento intrínseco obtenido del dataset de `imagenet`.

En eso consiste el **transfer learning**, en utilizar un modelo preentrenado para una tarea diferente a la que fue entrenado originalmente.

````python
learn.freeze()
learn.fine_tune(4)
learn.unfreeze()
````

````python
learn.fine_tune(10) 
````

La función `prob(letra)` obtiene la probabilidad de que la letra que se ha predicho sea la correcta.

````python
def prob(letter): 
    '''letter es un int, la posicion de la letra en el TensorBase'''
    valueLetter = predict[2][letter].item()
    tensorSum = 0
    for i in range (len(predict[2])):
        prediction = predict[2][i].item()
        tensorSum = tensorSum + prediction
    probLetter = valueLetter/tensorSum
    return int(probLetter*100)
````
